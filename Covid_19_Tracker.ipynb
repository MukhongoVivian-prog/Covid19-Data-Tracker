```python
import pandas as pd
```

```python
df = pd.read_csv('/content/covid.csv')
```

```python
countries_of_interest = ['Asia', 'Africa']
df_filtered = df[df['continent'].isin(countries_of_interest)]
```

```python
df_filtered.head()
```

```python
df_filtered1 = df_filtered.dropna(subset=['date', 'total_cases', 'total_deaths'])
```

```python
df_filtered1
```

```python
df_filtered1['date'] = pd.to_datetime(df_filtered1['date'])
```

```python
df_filtered1.info()
```

```python
df_filtered1['total_cases'] = df_filtered1['total_cases'].fillna(df_filtered1['total_cases'].mean())
```

```python
df_filtered1['total_deaths'] = df_filtered1['total_deaths'].fillna(df_filtered1['total_deaths'].mean())
```

EXPLORATORY DATA ANALYSIS

```python
import matplotlib.pyplot as plt

for country in countries_of_interest:
    country_data = df_filtered1[df_filtered1['continent'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.title('Total Cases Over Time for Selected Continents')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

for country in countries_of_interest:
    country_data = df_filtered1[df_filtered1['continent'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.title('Total Deaths Over Time for Selected Continents')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


for country in countries_of_interest:
    country_data = df_filtered1[df_filtered1['continent'] == country].copy() # Create a copy to avoid SettingWithCopyWarning
    country_data['daily_new_cases'] = country_data['total_cases'].diff()
    # Handle potential NaN values in the first row after diff()
    country_data['daily_new_cases'] = country_data['daily_new_cases'].fillna(0)

    plt.plot(country_data['date'], country_data['daily_new_cases'], label=country)

plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.title('Daily New Cases Over Time for Selected Continents')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Visualizing Vaccination Progress

```python
if 'total_vaccinations' in df_filtered1.columns:
    for country in countries_of_interest:
        country_data = df_filtered1[df_filtered1['continent'] == country].dropna(subset=['total_vaccinations']) # Drop rows with missing vaccination data

        if not country_data.empty:
            plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

    plt.xlabel('Date')
    plt.ylabel('Cumulative Vaccinations')
    plt.title('Cumulative Vaccinations Over Time for Selected Continents')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("The 'total_vaccinations' column was not found in the DataFrame.")
```

Compare % vaccinated population.

```python

# Check if necessary columns exist
if 'total_vaccinations' in df_filtered1.columns and 'population' in df_filtered1.columns:
    df_filtered1['vaccinated_percentage'] = (df_filtered1['total_vaccinations'] / df_filtered1['population']) * 100

    for country in countries_of_interest:
        # Filter data for the country and drop rows with missing percentage or population data
        country_data = df_filtered1[df_filtered1['continent'] == country].dropna(subset=['vaccinated_percentage', 'population'])

        if not country_data.empty: # Check if there's data after dropping NaNs
            plt.plot(country_data['date'], country_data['vaccinated_percentage'], label=country)

    plt.xlabel('Date')
    plt.ylabel('Percentage of Population Vaccinated')
    plt.title('Percentage of Population Vaccinated Over Time for Selected Continents')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Required columns ('total_vaccinations' and/or 'population') not found in the DataFrame.")
```

 Build a Choropleth Map

```python

latest_date = df_filtered1['date'].max()

latest_data_for_map = df_filtered1[df_filtered1['date'] == latest_date]

# Select the relevant columns for the map
map_data = latest_data_for_map[['iso_code', 'total_cases', 'continent']].copy()

# Drop rows with missing iso_code, as it's essential for mapping
map_data.dropna(subset=['iso_code'], inplace=True)

# You might want to aggregate data by iso_code in case there are multiple entries for the same day (though unlikely for latest date)
map_data = map_data.groupby('iso_code').agg({
    'total_cases': 'sum',
    'continent': 'first' # Keep the continent information
}).reset_index()

print("Prepared data for Choropleth map:")
print(map_data.head())
```

```python
import plotly.express as px

fig = px.choropleth(map_data,
                    locations="iso_code", # Column with ISO country codes
                    color="total_cases", # Column with the data to color the map
                    hover_name="continent", # Information to display when hovering over a continent
                    color_continuous_scale="Viridis", # Color scale to use
                    title="Total Cases by Continent (Latest Date)", # Title of the map
                    scope="world" # Set the scope of the map (world, USA, Europe, Asia, Africa, South America, North America)
                   )
fig.show()
```

# Anomalies and Observations in Africa and Asia

*   **Sudden Spikes in Daily Cases (Asia):** The daily new cases plot for Asia shows notable, sharp spikes at various points in time. These rapid increases could indicate periods of intense outbreaks, but also warrant considering whether they might be influenced by changes in testing volume or data reporting practices in specific countries within Asia.
*   **Consistent Trend in Cumulative Cases and Deaths (Africa vs. Asia):** While both continents show an overall increase in total cases and deaths over time, the *shape* of the curves differs. Africa's cumulative trends appear somewhat smoother compared to Asia, which exhibits more pronounced periods of accelerated growth in cases. This difference could reflect variations in testing, reporting consistency, or the timing and impact of waves across different regions.
*   **Vaccination Coverage Disparity:** The plot comparing the percentage of the population vaccinated over time clearly highlights a significant difference between Asia and Africa. Asia consistently shows a higher percentage of its population vaccinated throughout the analyzed period. This disparity is a key observation regarding the progress of vaccination campaigns in these two continents.
*   **Potential Data Gaps in Vaccination Data:** While visualizing cumulative vaccinations and vaccinated percentage, it was necessary to drop rows with missing data. If there are significant gaps in the vaccination data for certain periods or countries within Africa or Asia, this is an anomaly in the dataset itself that could affect the completeness of the vaccination plots.
*   **Impact of Policy or Seasonal Factors:** Observing the daily new cases plot, look for any cyclical patterns or sudden shifts in the trend that might align with major policy changes implemented in specific countries within these continents, or potentially seasonal effects on transmission.
